# Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
rel = sns.jointplot(data=df, x='sqft_living', y='price', kind='reg')
rel.fig.suptitle('Зависимость стоимости дома от жилой площади')
```

Мы видим что чем больше жилая площадь дома тем он дороже.

```python
plt.figure(figsize = (16, 8))
sns.scatterplot(x=df['price'], y=df['grade'])
plt.xlabel('Стоимость дома')
plt.ylabel('Количество спален')
plt.title('Зависимость цены дома от количества спален')
```

По графику видно что чем больше спален тем дороже дом.

```python
rel = sns.jointplot(data=df, x='yr_built', y='price', kind='scatter')
rel.fig.suptitle('Зависимость цены от года постройки дома')
```

Особой зависимости между годом постройки дома и ценой нет, но видно что новые дома немного дороже старых.

```python
df_renovated = df[['yr_renovated', 'price']]

def set_renovation(year):
    if year > 0:
        return 'Yes'
    else:
        return 'No'
    
df_renovated['yr_renovated'] = df_renovated['yr_renovated'].apply(set_renovation)

sns.barplot(data=df_renovated, x='yr_renovated', y='price')
```

Дома с реновацией стоят дороже.

```python
df_corr = df[['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'yr_built']]

corr_matrix = df_corr.corr()
corr_matrix = np.round(corr_matrix, 1)

sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm')

df_corr

```

По тепловой карте видно что есть коррелиция между жилой и общей площадью и между жилой площадью и ценой дома.
